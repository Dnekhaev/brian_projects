```python
# function to visualise the connectivity of Synapses S
def visualise_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    figure(figsize=(10, 4))
    subplot(121)
    plot(zeros(Ns), arange(Ns), 'ok', ms=10)
    plot(ones(Nt), arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        plot([0, 1], [i, j], '-k')
    xticks([0, 1], ['Source', 'Target'])
    ylabel('Neuron index')
    xlim(-0.1, 1.1)
    ylim(-1, max(Ns, Nt))
    subplot(122)
    plot(S.i, S.j, 'ok')
    xlim(-1, Ns)
    ylim(-1, Nt)
    xlabel('Source neuron index')
    ylabel('Target neuron index')
```

```python
# equations for input poisson neurons (added rates variable for each neuron, no voltage and current)
eqs_input_neuron = '''
da/dt = -a/alpha : 1
dtheta/dt = -theta/beta : 1
diff = a - c*theta : 1
rates : Hz
train : 1
'''
# equations for neurons of hidden layer
eqs_hidden_neuron = '''
dv/dt = (V0-v+I)/tau : 1 (unless refractory)
dI/dt = -I/(tau_I) : 1
da/dt = -a/alpha : 1
dtheta/dt = -theta/beta : 1
V0 :1
diff = a - c*theta : 1
train : 1

dhold_hidden/dt = (20-hold_hidden)/(15*ms) : 1
'''

eqs_output_neuron = '''
dv/dt = (V0-v+I)/tau : 1 (unless refractory)
dI/dt = -I/(tau_I) : 1
da/dt = -a/alpha : 1
dtheta/dt = -theta/beta : 1
V0 :1
diff = a - c*theta : 1
train : 1

dhold_output/dt = (1-hold_output)/(15*ms) : 1
'''


# each forward connection synapse has it's' own weight 'w'
eqs_syn = '''
w : 1
'''

# each negative connection synapse has it's own weight 'w'
eqs_syn_neg = '''
w_neg: 1
'''

# equations that describe changes if presynaptic spike of the forward-riented synapse of input layer occures
eqs_input_pre = '''
I_post += w
a_pre += 1./n_hidden *1*ms/(alpha)
theta_pre += 1./n_hidden *1*ms/(beta)
'''

# equations that describe changes if presynaptic spike of the forward synapse of hidden layer occures
eqs_hidden_pre = '''
I_post += w
a_pre += 1./classes * 1*ms/alpha
theta_pre += 1./classes * 1*ms/beta
'''
# equations that describe changes if postsynaptic spike of the forward synapse occures
eqs_post = '''
w = clip(w + train_pre * (-0.001 + lr*diff_pre), 0, wmax)
'''

# equations that describe changes if postsynaptic spike of the forward synapse of hidden layer occures
eqs_hidden_post = '''
w = clip(w + train_pre * (-0.001 + lr*diff_pre), 0, wmax)
a_post += 10./n_hidden * 1*ms/alpha
theta_post += 10./n_hidden * 1*ms/beta
'''

#w = clip(w +(-lr*log(1+exp(-3*(w+0.1)*diff_pre))+0.05)*(wmax-w), 0, wmax)

# equations that describe changes if presunaptic spike of the negative synapse occures
eqs_neg_pre = '''
I_post -= 150./(n_hidden-1)*w_neg
w_neg = clip(w_neg + train_pre * (lr_n*diff_post), 0, wmax)
'''

eqs_reverse_pre = '''
I_post += w
w = clip(w + train_pre * (lr*diff_post), 0, wmax)
'''

reset_hidden = '''
v = 0
hold_hidden += 2
'''

reset_output = '''
v = 0
hold_output += 0.1
'''
```

```python
P = NeuronGroup(n_input, eqs_input_neuron, method=linear, 
                        threshold='rand()<rates*dt', dt = time_step)
P.a = np.random.random(n_input)
P.theta = np.random.random(n_input)
# train is parameter that allows synaptic weights to change (if train==1) or disables weight change (if train==0)
P.train = np.ones_like(P.train)

G = NeuronGroup(n_hidden, eqs_hidden_neuron, method=linear, threshold='v > hold_hidden', reset=reset_hidden, 
                refractory=2*ms, dt=time_step)
G.train = np.ones_like(G.train)
G.hold_hidden = np.ones_like(G.hold_hidden)*20.

InputConnection = Synapses(P, G, eqs_syn, on_pre=eqs_input_pre, on_post=eqs_post, dt=time_step)
InputConnection.connect()
InputConnection.w = np.random.random(n_input*n_hidden)
        
NegativeConnection = Synapses(G, G, eqs_syn_neg, on_pre=eqs_neg_pre, dt=time_step)
NegativeConnection.connect(condition='i != j')
NegativeConnection.w_neg = np.ones((n_hidden_height*n_hidden_width-1)*n_hidden_height*n_hidden_width) * 0.5
        
H = NeuronGroup(classes, eqs_output_neuron, method=linear, threshold='v > hold_output', reset=reset_output, refractory=2*ms, dt=time_step)
H.train = np.ones_like(H.train)
H.hold_output = np.ones_like(H.hold_output)*1.

OutputConnection = Synapses(G, H, eqs_syn, on_pre=eqs_hidden_pre, on_post=eqs_hidden_post, dt=time_step)
OutputConnection.connect()
OutputConnection.w = np.random.random(n_hidden*classes)

ReverseConnection = Synapses(H, G, eqs_syn, on_pre=eqs_reverse_pre, dt=time_step)
ReverseConnection.connect()
ReverseConnection.w = np.random.random(n_hidden*classes)
```

```python
def accuracy(y, X_norm, x_mean):
    pred = []
    for XX in X_norm:
        sver = []
        for filtr in x_mean:
            sver.append(np.sum(filtr*XX))
        pred.append(np.argmax(sver))

    yyy =[yy[0] for yy in y]
    return np.sum([xx==yy for xx, yy in zip(pred,y)])/float(len(pred))*100
```

```python
import random

def annealing_optimize(initial, X_norm, x_mean, costf, T=0.01, cool=0.99, step=0.005):
    # Выбрать случаиное решение
    vec = initial

    while T > 0.0001:
        # Выбрать один из индексов
        i = random.randint(0, 9)

        # Выбрать направление изменения
        q = random.random()-0.5
        if q>0:
            q=1
        else:
            q=-1
        
        direction = step*q

        # Создать новыи список, в котором одно значение изменено
        vecb = vec[:]
        vecb[i] += direction
        if vecb[i] < 0.: vecb[i] = 0.
        elif vecb[i] > 3.: vecb[i] = 3.

        # Вычислить текущую и новую стоимость
        ea = 100.-costf(y, X_norm, [v*xx for v,xx in zip(vec, x_mean)])
        eb = 100.-costf(y, X_norm, [v*xx for v,xx in zip(vecb, x_mean)])
        p = pow(math.e, (-(eb-ea))/T)

        # Новое решение лучше? Если нет, метнем кости
        message = 0
        if (eb < ea):
            vec = vecb
            message = 1
        elif random.random() < p:
            vec = vecb
            message = 2

        # Уменьшить температуру
        print 'T={}, mistake_pre={} %, mistake_post={}%, proba={}, change={}'.format(round(T,2), ea, eb, round(p,2), message)
        T = T * cool
    return vec, eb

result2, score = annealing_optimize(result, X_norm, x_mean, accuracy)
print result2, score
```
