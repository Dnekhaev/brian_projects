```python
time_step = 0.5*ms #timestep for simulation

n_input_width = 9 #
n_input_height = 9 #
n_input = float(n_input_height*n_input_width)
n_hidden_width = 6#
n_hidden_height = 6 #
n_hidden = float(n_hidden_height*n_hidden_width)
lr=0.1 # learning rate for forward connections
lr_n=1.0 # learning rate for negative intrinsic connections
lr_rev = 1.0 # learning rate for reverse connections
runtime = 100*ms # simulation time per one image sample
tau = 10*ms # tau for neuron's voltage 'v'
tau_I = 15*ms # tau for neuron's current 'I'
#hold_hidden = 20 # theshold for neuron's voltage to generate a spike
#hold_output = 1.0 # theshold for neuron's voltage to generate a spike
c = 1. # coefficient to calculate difference between rate and average rate: difference = rate - c * average_rate
coef = 1. #actually don't remember about this parameter, but if it is equal to one, it doesn't change anything
alpha = 30*ms #tau for computing exponential mean of spike-array to calculate rate 'a'
beta= 500*ms #tau for computing exponential mean of spike-array to calculate average_rate 'theta'
wmax=1.0 #maximum weight value

hold_h = 3.0
hold_o = 4.0
c_inp = 8.0
c_neg = 45.0
c_rev = 2.0
c_out = 3.0
```

```python
# equations for input poisson neurons (added rates variable for each neuron, no voltage and current)
eqs_input_neuron = '''
da/dt = -a/alpha : 1
dtheta/dt = -theta/beta : 1
diff = a - c*theta : 1
rates : Hz
train : 1
'''
# equations for neurons of hidden layer
eqs_hidden_neuron = '''
dv/dt = (V0-v+I)/tau : 1 (unless refractory)
dIinp/dt = -Iinp/(tau_I) : 1
dIneg/dt = -Ineg/(tau_I) : 1
dIrev/dt = -Irev/(tau_I) : 1
I = Iinp + Ineg + Irev :1
da/dt = -a/alpha : 1
dtheta/dt = -theta/beta : 1
V0 :1
diff = a - c*theta : 1
train : 1

dhold_hidden/dt = (hold_h-hold_hidden)/(15*ms) : 1
'''

eqs_output_neuron = '''
dv/dt = (V0-v+I)/tau : 1 (unless refractory)
dI/dt = -I/(tau_I) : 1
da/dt = -a/alpha : 1
dtheta/dt = -theta/beta : 1
V0 :1
diff = a - c*theta : 1
train : 1

dhold_output/dt = (hold_o-hold_output)/(15*ms) : 1
'''


# each forward connection synapse has it's' own weight 'w'
eqs_input_syn = '''
w : 1
'''

eqs_output_syn = '''
w : 1
w_rev : 1
'''

# each negative connection synapse has it's own weight 'w'
eqs_syn_neg = '''
w_neg: 1
'''

# equations that describe changes if presynaptic spike of the forward-riented synapse of input layer occures
eqs_input_pre = '''
Iinp_post += w / c_inp
a_pre += 1./n_hidden *1*ms/(alpha)
theta_pre += 1./n_hidden *1*ms/(beta)
'''

# equations that describe changes if postsynaptic spike of the forward synapse occures
eqs_input_post = '''
w = clip(w + train_pre * (-0.001 + lr*diff_pre), 0, wmax)
'''

# equations that describe changes if presynaptic spike of the forward synapse of hidden layer occures
eqs_output_pre = '''
I_post += w / c_out
a_pre += 1./classes * 1*ms/alpha
theta_pre += 1./classes * 1*ms/beta
'''

# equations that describe changes if postsynaptic spike of the forward synapse of hidden layer occures
eqs_output_post = '''
a_post += 1./n_hidden * 1*ms/alpha
theta_post += 1./n_hidden * 1*ms/beta
w = clip(w + train_pre * (-0.001 + lr*diff_pre), 0, wmax)
Irev_pre += w_rev / c_rev
w_rev = clip(w_rev + train_post * lr_rev * w * diff_pre , 0, wmax)
'''

# equations that describe changes if spike of the negative synapse occures
eqs_neg_pre = '''
Ineg_post += w_neg / c_neg
'''

eqs_neg_post = '''
w_neg = clip(w_neg + train_pre * (lr_n*diff_pre), -wmax, wmax)
'''

#eqs_reverse_pre = '''
#I_post += w
#w = clip(w + train_pre * (lr*diff_post), 0, wmax)
#'''

reset_hidden = '''
v = 0
hold_hidden += 0.1*hold_h
'''

reset_output = '''
v = 0
hold_output += 0.1*hold_o
'''
```

```python
class SpikingNN():
    def __init__(self, n_input_height = 28, n_input_width = 28, n_hidden_width = 10, n_hidden_height = 10, 
                 lr=0.1, lr_n=1.0, lr_rev=0.1, runtime = 100*ms,
                 tau = 10*ms, tau_I = 15*ms, c = 1., coef = 1., alpha = 30*ms, beta= 500*ms,
                 hold_h = 3.0, hold_o = 4.0, c_inp = 8.0, c_neg = 45.0, c_rev = 2.0, c_out = 3.0,
                 eqs_input_neuron = eqs_input_neuron, eqs_hidden_neuron = eqs_hidden_neuron, 
                 eqs_output_neuron = eqs_output_neuron,
                 eqs_input_syn = eqs_input_syn, eqs_syn_neg = eqs_syn_neg, eqs_output_syn = eqs_output_syn,
                 eqs_input_pre = eqs_input_pre, eqs_input_post = eqs_input_post, 
                 eqs_output_pre = eqs_output_pre, eqs_output_post = eqs_output_post, 
                 eqs_neg_pre = eqs_neg_pre, eqs_neg_post = eqs_neg_post,
                 reset_hidden = reset_hidden, reset_output = reset_output):
        
        # set all parameters to the object variables
        self.n_input_width = n_input_width
        self.n_input_height = n_input_height
        self.n_input = n_input_height * n_input_width
        self.n_hidden_width = n_hidden_width
        self.n_hidden_height = n_hidden_height
        self.n_hidden = n_hidden_height * n_hidden_width
        self.lr = lr
        self.lr_n = lr_n
        self.lr_rev = lr_rev
        self.c = c
        
        self.eqs_input_neuron = eqs_input_neuron
        self.eqs_hidden_neuron = eqs_hidden_neuron
        self.eqs_output_neuron = eqs_output_neuron
        
        self.eqs_input_syn = eqs_input_syn
        self.eqs_output_syn = eqs_output_syn
        self.eqs_syn_neg = eqs_syn_neg
        
        self.eqs_input_pre = eqs_input_pre
        self.eqs_input_post = eqs_input_post
        
        self.eqs_output_pre = eqs_output_pre
        self.eqs_output_post = eqs_output_post
        
        self.eqs_neg_pre = eqs_neg_pre
        self.eqs_neg_post = eqs_neg_post
        
        self.reset_hidden = reset_hidden
        self.reset_output = reset_output
        
        self.wmax = 1.
        self.coef = coef
        self.runtime = runtime*self.coef
        self.tau = tau*self.coef
        self.tau_I = tau_I*self.coef
        self.alpha = alpha*self.coef
        self.beta = beta*self.coef
        
        self.hold_h = hold_h
        self.hold_o = hold_o
        self.c_inp = c_inp
        self.c_neg = c_neg
        self.c_rev = c_rev
        self.c_out = c_out

        # array of predictions
        self.pred = []
        
        
    def architecture(self):
        
        # Create Neurons and Connections between them.
        # P -> InputConnection -> G -> OutputConnection <----> H (N = num classes) 
        #                         |
        #                 NegativeConnection
        #                         |
        #                         G
        #
        
        P = NeuronGroup(self.n_input, self.eqs_input_neuron, method=linear, 
                        threshold='rand()<rates*dt', dt = time_step)
        P.a = np.random.random(self.n_input)
        P.theta = np.random.random(self.n_input)
        
        # train is parameter that allows synaptic weights to change (if train==1) or disables weight change (if train==0)
        P.train = np.ones_like(P.train)
        
        G = NeuronGroup(self.n_hidden, self.eqs_hidden_neuron, method=linear, threshold='v > hold_hidden', 
                        reset=self.reset_hidden, refractory=2*ms, dt=time_step)
        G.train = np.ones_like(G.train)
        InputConnection = Synapses(P, G, self.eqs_input_syn, on_pre=self.eqs_input_pre, on_post=self.eqs_input_post, dt=time_step)
        
        InputConnection.connect()
        InputConnection.w = np.random.random(self.n_input*self.n_hidden)
        
        NegativeConnection = Synapses(G, G, self.eqs_syn_neg, on_pre=self.eqs_neg_pre, on_post=self.eqs_neg_post,
                                      dt=time_step)
        NegativeConnection.connect(condition='i != j')
        NegativeConnection.w_neg = np.zeros_like(NegativeConnection.w_neg)
        
        H = NeuronGroup(classes, self.eqs_output_neuron, method=linear, threshold='v > hold_output', reset=self.reset_output,
                        refractory=2*ms, dt=time_step)
        H.train = np.ones_like(H.train)
        
        OutputConnection = Synapses(G, H, self.eqs_output_syn, on_pre=self.eqs_output_pre, on_post=self.eqs_output_post, dt=time_step)
        OutputConnection.connect()
        OutputConnection.w = np.random.random(classes*self.n_hidden)
        OutputConnection.w_rev = np.zeros_like(OutputConnection.w_rev)
        
        return P, G, InputConnection, NegativeConnection, H, OutputConnection
    
    def pp(self, s=0, d=0, w = 1, h = 1):
        return [mpatches.Rectangle((i+d*(self.n_input_width+1), j+s*(self.n_input_height+1)), 1, 1) 
                for j in np.arange(h-1, -1, -1) for i in np.arange(w)]

    def create_canvas(self):
        back_color = "yellow"
        width = self.n_input_width*self.n_hidden_width + self.n_hidden_width-1
        height = self.n_input_height*self.n_hidden_height + self.n_hidden_height-1

        self.fig, (self.ax, self.ax2, self.ax3) = plt.subplots(1, 3, figsize=(15,5))
        self.ax.set(xlim=[0, width], ylim=[0, height]) # Or use "ax.axis([x0,x1,y0,y1])"
        self.ax2.set(xlim=[0, self.n_hidden_width], ylim=[0, self.n_hidden_height]) # Or use "ax.axis([x0,x1,y0,y1])"
        self.ax3.set(xlim=[0, self.n_input_width], ylim=[0, self.n_input_height]) # Or use "ax.axis([x0,x1,y0,y1])"
        self.fig.canvas.draw()

        self.patches = []
        self.patches2 = []
        self.patches3 = []
        self.patches = [self.pp((self.n_hidden_height-1-s),d, self.n_input_width, self.n_input_height) for s in np.arange(self.n_hidden_height) 
                        for d in np.arange(self.n_hidden_width)]
        self.patches2 = self.pp(0,0, self.n_hidden_width, self.n_hidden_height)
        self.patches3 = self.pp(0,0, self.n_input_width, self.n_input_height)
        
        for s in np.arange(len(self.patches)):
            for j in np.arange(len(self.patches[0])):
                self.ax.add_artist(self.patches[s][j])
                
        for j in np.arange(len(self.patches2)):
            self.ax2.add_artist(self.patches2[j])
            
        for j in np.arange(len(self.patches3)):
            self.ax3.add_artist(self.patches3[j])
        
    def update(self, P, G, InputConnection, NegativeConnection, H, OutputConnection, k, rate, SMP, SMG, SMH):
        
        lr = self.lr
        lr_n = self.lr_n
        lr_rev = self.lr_rev
        c = self.c
        tau = self.tau
        tau_I = self.tau_I
        alpha = self.alpha 
        beta = self.beta
        wmax = self.wmax
        n_hidden = float(self.n_hidden)
        n_input = float(self.n_input)
        
        hold_h = self.hold_h
        hold_o = self.hold_o
        c_inp = self.c_inp
        c_neg = self.c_neg
        c_rev= self.c_rev
        c_out = self.c_out

        P.rates = 250 * self.coef * rate * Hz
        beg_train = clock()
        run(self.runtime)
        end_train = clock()
        print '   time per train: {}, simulation time {}'.format(end_train-beg_train, self.runtime)
        rate_z = np.zeros(n_input)
        P.rates = rate_z*Hz
        beg_holdout = clock()
        run(self.runtime*0.3)
        end_holdout = clock()
        print '   time per holdout: {}, simulation time {}'.format(end_holdout-beg_holdout, self.runtime*0.3)

        colors = np.zeros((self.n_hidden, self.n_input))
        for s in np.arange(len(self.patches)):
            for j in np.arange(len(self.patches[0])):
                colors[s][j]=InputConnection.w[s+j*int(self.n_hidden)]
                self.patches[s][j].set_facecolor(np.array([colors[s][j], colors[s][j], colors[s][j]]))
        
        colors_neg = np.zeros((self.n_hidden))
        
        for j in np.arange(self.n_hidden):        
            colors_neg[j]=NegativeConnection.w_neg[j*(int(self.n_hidden)-1)]
            self.patches2[j].set_facecolor(np.array([colors_neg[j], colors_neg[j], colors_neg[j]]))
            
        for j in np.arange(len(self.patches3)):
            self.patches3[j].set_facecolor(np.array([rate[j], rate[j], rate[j]]))

        self.fig.canvas.blit(self.ax.bbox)
        self.fig.canvas.blit(self.ax2.bbox)
        self.fig.canvas.blit(self.ax3.bbox)
        
        self.fig.canvas.draw()
    
    def train(self, X, y, epochs_train = 0, epochs_test = 0, show = True, classifier = False, train = True, 
              pretrain = False, cheat = False, train_H = False):
        
        if cheat:
            P, G, InputConnection, NegativeConnection, H, OutputConnection = self.architecture()
            
            dig_y = y[:self.n_hidden]
            dig_X = X[:self.n_hidden]
            for s in np.arange(self.n_hidden):
                for j in np.arange(self.n_input):
                    InputConnection.w[s+j*self.n_hidden] = dig_X[s][j]
            
            NegativeConnection.w_neg = np.zeros_like(NegativeConnection.w_neg)
            if show:
                plt.show()
            
            self.P = P
            self.G = G
            self.InputConnection = InputConnection
            self.NegativeConnection = NegativeConnection
            self.OutputConnection = OutputConnection
            self.H = H
            
            colors = np.zeros((self.n_hidden, self.n_input))
            for s in np.arange(len(self.patches)):
                for j in np.arange(len(self.patches[0])):
                    colors[s][j]=InputConnection.w[s+j*self.n_hidden]
                    self.patches[s][j].set_facecolor(np.array([colors[s][j], colors[s][j], colors[s][j]]))

            self.fig.canvas.blit(self.ax.bbox)
        
            self.fig.canvas.draw()
            
        if train:
            if pretrain:
                P = self.P
                G = self.G
                InputConnection = self.InputConnection
                NegativeConnection = self.NegativeConnection
                OutputConnection = self.OutputConnection
                H = self.H
                
                SMP = self.SMP
                SMG = self.SMG
                SMH = self.SMH
                
                P.train = np.ones_like(P.train)
                G.train = np.ones_like(G.train)
                H.train = np.zeros_like(H.train)
                if train_H:
                    H.train = np.ones_like(H.train)
                else: 
                    H.train = np.zeros_like(H.train)
            
            else:
                P, G, InputConnection, NegativeConnection, H, OutputConnection = self.architecture()
                SMP = StateMonitor(P, ['a'], record = True)
                SMG = StateMonitor(G, ['I', 'Iinp', 'Ineg', 'Irev','a'], record = True)
                SMH = StateMonitor(H, ['I', 'a'], record = True)
                
                if train_H:
                    H.train = np.ones_like(H.train)
                else: 
                    H.train = np.zeros_like(H.train)
            k = 0
            
            #SMC = SpikeMonitor(G[], record = True)
                
            if show:
                plt.show()
            
            while k < epochs_train:
                rate =  X[k%(X.shape[0])].reshape(self.n_input)
                beg_update = clock()
                self.update(P, G, InputConnection, NegativeConnection,  H, OutputConnection, k, rate, SMP, SMG, SMH)
                end_update = clock()
                print 'image {}, time per update function: {}'.format(k, end_update-beg_update)
                k += 1

            self.P = P
            self.G = G
            self.InputConnection = InputConnection
            self.NegativeConnection = NegativeConnection
            self.OutputConnection = OutputConnection
            self.H = H
            self.SMP = SMP
            self.SMG = SMG
            self.SMH = SMH
        
        if classifier:
            P, G, InputConnection, NegativeConnection, H, OutputConnection = self.create_classifier(
                P, G, InputConnection, NegativeConnection, H, OutputConnection, X, y)
            

            SC0 = SpikeMonitor(G, record=True)
            self.SC0 = SC0
            StateMH = StateMonitor(H, ['v', 'I', 'a'], record=True)
            self.StateMonitorH = StateMH
            StateMG = StateMonitor(G, ['v', 'I', 'Iinp', 'Ineg', 'Irev', 'a'], record=True)
            self.StateMonitorG = StateMG

            lr = self.lr
            lr_n = self.lr_n
            lr_rev = self.lr_rev
            c = self.c
            tau = self.tau
            tau_I = self.tau_I
            alpha = self.alpha 
            beta = self.beta
            wmax = self.wmax
            
            hold_h = self.hold_h
            hold_o = self.hold_o
            c_inp = self.c_inp
            c_neg = self.c_neg
            c_rev= self.c_rev
            c_out = self.c_out

            k=0
            print 'testing'
            while k < epochs_test:

                rate =  X[(k)%(X.shape[0])].reshape(self.n_input)
                P.rates = 250 * self.coef * rate * Hz
                start = clock()
                run(self.runtime)
                end = clock()
                
                self.pred.append(np.argmax(self.StateMonitorH.a[:,-1]))
                accuracy = self.accuracy(y[:len(self.pred)], self.pred)
                
                if k % 1 == 0:
                    print k, accuracy, '%'
                    
                print '{} runtime for {} classification test'.format(end - start, self.runtime)
                print 'w_neg min {}, w_neg max {}'.format(min(self.NegativeConnection.w_neg), max(self.NegativeConnection.w_neg))
                print 'w_output min {}, w_output max {}'.format(min(self.OutputConnection.w), max(self.OutputConnection.w))
                print 'w_rev min {}, w_rev max {}'.format(min(self.OutputConnection.w_rev), max(self.OutputConnection.w_rev))
                
                P.rates = np.zeros_like(P.rates)
                
                #G.v = np.zeros_like(G.v)
                #G.I = np.zeros_like(G.I)
                
                #H.a = np.zeros_like(H.a)
                #H.theta = np.zeros_like(H.theta)
                #H.v = np.zeros_like(H.v)
                #H.I = np.zeros_like(H.I)
                run(self.runtime)
                k += 1
                
        self.P = P
        self.G = G
        self.InputConnection = InputConnection
        self.NegativeConnection = NegativeConnection
        self.OutputConnection = OutputConnection
        self.H = H
            
    def create_classifier(self, P, G, InputConnection, NegativeConnection, H,
                         OutputConnection, X, y):
        
        
        # xx first dimension is 10 (classes). Contains images of digits grouped by label
        xx = []

        for j in np.arange(classes):
            xs = []
            for i in np.arange(len(y)):
                if y[i] == j:
                    xs.append(X[i])
            xx.append(xs)
            
        # x_mean contains the mean images per class label
        x_mean = []
        for i in np.arange(len(xx)):
            x_mean.append(np.mean(np.array(xx[i]), axis=0))

        colors = np.zeros((self.n_hidden, self.n_input))
        for s in np.arange(len(self.patches)):
            for j in np.arange(len(self.patches[0])):
                colors[s][j]=net.InputConnection.w[s+j*self.n_hidden]
                
        # dim (10, n_hidden), convoltions of all hidden weight images with 10 class images_mean
        scals = []

        for k in np.arange(classes):
            scal = []
            for i in np.arange(self.n_hidden):
                #contains convolution of mean class image [k] with particular hidden weight image
                scal.append(np.sum(np.array(colors[i]*x_mean[k])))
            #convolutions of all hidden weight images with [k] class
            scals.append(scal)
        
        scals = np.array(scals)
        
        # dim n_hidden, "labels" for each hidden neuron
        targets = np.argmax(scals, axis = 0)
        self.targets = targets
        
        P.train = np.zeros_like(P.train)
        G.train = np.ones_like(G.train)
        H.train = np.ones_like(H.train)
        
        OutputConnection.w = np.zeros_like(OutputConnection.w)
        OutputConnection.w_rev = np.zeros_like(OutputConnection.w_rev)
        
        #CHEEEEAAAAT
        #targets = y
        #self.targets = targets
        for s in np.arange(self.n_hidden):
            target = targets[s]
            OutputConnection.w[int(target + s * (classes))] = 0.9
            OutputConnection.w_rev[int(target + s* (classes))] = 0.9
        
        
        return P, G, InputConnection, NegativeConnection, H, OutputConnection
        
    def accuracy(self, y, pred):
        accuracy = 0

        for a, b in zip(y, pred):
            if a == b:
                accuracy += 1

        accuracy /= float(len(pred))
        return round(accuracy*100, 2)  
```
